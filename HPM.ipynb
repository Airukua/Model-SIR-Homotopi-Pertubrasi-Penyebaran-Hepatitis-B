```python
S=1847950
I=569
R=569

# Total populasi
N = 1849037

s0= 0.9994121264204016
i0= 0.00030772775233810896
r0= 0.00030772775233810896

t=[]

for i in range(1,21) :
   x = i/100
   t.append(x)
```

```python
#nilai parameter
k = 0.00168
miu1 = 0.01513
miu2 = 0
beta = 0.00031
alpha = 1
```

```python
def v1(x):
    v = [x]
    for i in range(3):
        v_new = k - (miu1 * v[i-1]) - beta * v[i-1]
        v.append(v_new)
        x = v_new
        total = sum(v)
    return total
```

```python
def v2(y):
    v = [y]
    for i in range(3):
        v_new = beta * v[i] - (miu1 + miu2) * v[i] - alpha * v[i]
        v.append(v_new)
        y = v_new
        total = sum(v)
    return total
```

```python
def v3(z):
    v = [z]
    for i in range(3):
        v_new = alpha*v[i]-miu1*v[i]
        v.append(v_new)
        z = v_new
        total = sum(v)
    return total
```

```python
x = s0
s = [x]

for i in range(19):
    s_new = v1(s[i])
    s.append(s_new)
    x = s_new
```

```python
y = i0
I = [y]

for i in range(19):
    I_new = v2(I[i])
    I.append(I_new)
    y = I_new
```

```python
z = r0
R = [z]

for i in range(19):
    R_new = v3(R[i])
    R.append(R_new)
    y = R_new
```
